```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(data=Air, x="airline", y="price", estimator="mean")
plt.xticks(rotation=45)
plt.title("Average Price by Airline")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=Air, x="source_city", order=Air['source_city'].value_counts().index)
plt.title("Number of Flights per Source City")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=Air, x="class")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=Air, x="stops",y="price", estimator ="mean")
plt.title("Average Price by Number of stops")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=Air, x="price", bins =30,kde=True)
plt.title("Duration vs Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=Air, x="duration", y="price", hue="class")
plt.title("Duration vs Price")
plt.show()
```

```python
class_counts = Air["class"].value_counts()
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Ticket Class Distribution")
plt.show()
```
